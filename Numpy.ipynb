```python
import numpy as np
a=np.arange(12)
print(a)
print("Last element : ",a[-1])
print("Dimension of a : ",a.ndim)
print("Shape of a : ",a.shape)
print("Data type of each element : ",a.dtype)
```

```python
print(a)
print("##############")
b=a.reshape(3,4)
print(b)
print("Dimension of b : ",b.ndim)
print("Sum of all the elements : ",b.sum())
print("Average of all the elements : ",b.mean())
print("Row sum : ",b.sum(axis=1))
print("Column sum : ",b.sum(axis=0))
```

```python
print(b)
print("#############################")
print(b.T)
print("#############################")
print(np.ravel(b))
```
