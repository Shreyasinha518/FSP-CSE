```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/ML Folder/Mall_Customers (1).csv")
df
```

```python

x=df.iloc[:,[3,4]].values
x
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
sch.dendrogram(sch.linkage(x,method='ward'))
plt.xlabel("customers")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=5,random_state=0)
y_pred=model.fit_predict(x)
y_pred
```

```python
model.cluster_centers_
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],c='red',s=100,label="cluster0")
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],c='blue',s=100,label="cluster1")
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],c='green',s=100,label="cluster2")
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],c='cyan',s=100,label="cluster3")
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],c='magenta',s=100,label="cluster4")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c="black",s=100,marker="*",label="center")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()




```
