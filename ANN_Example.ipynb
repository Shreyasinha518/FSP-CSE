```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/ML Folder/Churn_Modelling (2).csv")
df
```

# New section

```python
x=df.iloc[:,3:-1].values
x
```

```python
from sklearn.preprocessing  import LabelEncoder
en1=LabelEncoder()
x[:,1]=en1.fit_transform(x[:,1])
en2=LabelEncoder()
x[:,2]=en2.fit_transform(x[:,2])
x

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn .model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train
```

```python
x_train.shape
```

```python

import keras
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(units=64,activation='relu',input_dim=10))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units=16,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()

```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32,validation_data=(x_test,y_test))
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred=model.predict(x_test)
y_pred=y_pred>.5
y_pred
```

```python
y_pred=model.predict(x_test)
y_pred=(y_pred>.5).astype("int32")
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy=",accuracy_score(y_test,y_pred))
print("Precission=",precision_score(y_test,y_pred))
print("Accuracy=",recall_score(y_test,y_pred))

```
